```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report,confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
training_data = pd.read_csv("D:\Modules_UCD\EEEN40690\Final project\Wine_train.csv")
training_data = training_data.drop('id',axis=1)
training_data
```

```python
test_data = pd.read_csv("D:\Modules_UCD\EEEN40690\Final project\Wine_test.csv")
test_data = test_data.drop('id',axis=1)
test_data
```

```python
X = training_data.drop('target', axis = 1)
y = training_data['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)

ss = StandardScaler()

X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
SVMlinear=SVC(kernel='linear', random_state = 0)
SVMpoly=SVC(kernel='poly', random_state=0)
SVMrbf=SVC(kernel='rbf', random_state=0)
SVMsigmoid=SVC(kernel='sigmoid', random_state=0)
```

```python
SVMlinear.fit(X_train,y_train)
SVMpoly.fit(X_train,y_train)
SVMrbf.fit(X_train,y_train)
SVMsigmoid.fit(X_train,y_train)
```

```python
SVMlinear_pred = SVMlinear.predict(X_test)
SVMpoly_pred = SVMpoly.predict(X_test)
SVMrbf_pred = SVMrbf.predict(X_test)
SVMsigmoid_pred = SVMsigmoid.predict(X_test)
```

```python
Accuracy_of_SVMlinear = accuracy_score(y_test, SVMlinear_pred)
print(Accuracy_of_SVMlinear)
Accuracy_of_SVMpoly = accuracy_score(y_test, SVMpoly_pred)
print(Accuracy_of_SVMpoly)
Accuracy_of_SVMrbf = accuracy_score(y_test, SVMrbf_pred)
print(Accuracy_of_SVMrbf)
Accuracy_of_SVMsigmoid = accuracy_score(y_test, SVMsigmoid_pred)
print(Accuracy_of_SVMsigmoid)
```

```python
test_data1 = ss.fit_transform(test_data)
SVMlinear_pred2 = SVMlinear.predict(test_data1)
SVMpoly_pred2 = SVMpoly.predict(test_data1)
SVMrbf_pred2 = SVMrbf.predict(test_data1)
SVMsigmoid_pred2 = SVMsigmoid.predict(test_data1)
```

```python
print("\n\nClassification Report for Linear SVM\n ",classification_report(y_test,SVMlinear_pred))


con_mat = confusion_matrix(y_test,SVMlinear_pred)
sns.heatmap(con_mat, annot=True, square=True)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Confusion Matrix')
```
